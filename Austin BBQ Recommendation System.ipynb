## Method 1: Word Frequency

### Importance

```python
def word_frequency(sentence):
    sentence =" ".join(sentence)
    #Create tokens
    new_words = word_tokenize(sentence)
    new_words = [t.lower() for t in new_words]
    
    #Remove stop words
    new_words =[t for t in new_words if t not in stopwords.words('english')]
    new_words = [t for t in new_words if t.isalpha()]
    
    #Lemmatization
    lemmatizer = WordNetLemmatizer()
    new_words =[lemmatizer.lemmatize(t) for t in new_words]
    
    #Counts words, pairs and trigrams
    counted = Counter(new_words)
    counted_2= Counter(ngrams(new_words,2))
    counted_3= Counter(ngrams(new_words,3))
    
    #Create dataframes
    word_freq = pd.DataFrame(counted.items(),columns=['word','frequency']).sort_values(by='frequency',ascending=False)
    word_pairs =pd.DataFrame(counted_2.items(),columns=['pairs','frequency']).sort_values(by='frequency',ascending=False)
    trigrams =pd.DataFrame(counted_3.items(),columns=['trigrams','frequency']).sort_values(by='frequency',ascending=False)
    return word_freq, word_pairs, trigrams
```

```python
words = pd.DataFrame()
bigrams = pd.DataFrame()
trigrams = pd.DataFrame()
bbq_list = data['BBQ'].unique()

for i in bbq_list:
    bbq_df = data[data['BBQ'] == i]
    bbq_1, bbq_2, bbq_3 = word_frequency(bbq_df['ReviewDescription'])
    bbq_1['BBQ'] = i
    bbq_2['BBQ'] = i
    bbq_3['BBQ'] = i
    words = words.append(bbq_1)
    bigrams = bigrams.append(bbq_2)
    trigrams = trigrams.append(bbq_3)

print(len(words))
print(len(bigrams))
print(len(trigrams))
```

```python
#Using textblob to identify parts of speech
from textblob import TextBlob

#Creating an empty column in tokens for part of speech
words['poc'] = ''

#Creating a loop to identify the parts of speech for each word
for i in range(len(words)):
    blob = TextBlob(words['word'][i])
    words.at[i, 'poc'] = blob.tags[0][1]
```

```python
#NN (nouns) represent the features for the beers
bbq_feat = words[words['poc'] == 'NN']
bbq_feat.head(30)
```

```python
#Getting the importance of each feature for that particular beer

bbq_total = bbq_feat.groupby(['BBQ']).sum()#Getting total words for each beer
bbq_join = pd.merge(bbq_feat, bbq_total, on = 'BBQ')
bbq_join = bbq_join.drop(['index_x', 'index_y'], axis = 1)#drop not needed indexes

#Getting the importance - Importance is the frequency of that word divided by the overall words for that beer
bbq_join['imp'] = pd.to_numeric(bbq_join['frequency_x']) / pd.to_numeric(bbq_join['frequency_y'])
```

```python
#Input 3 features and choose the best bbq based on the features
#The best bbq will be one for which the total importance for all 3 features will be maximum

attr1 = input("Enter your first attribute: ")
attr2 = input("Enter your second attribute: ")
attr3 = input("Enter your third attribute: ")

bbq_attribute = bbq_join[bbq_join['word'] == attr1].append(bbq_join[bbq_join['word'] == attr2]).append(bbq_join[bbq_join['word'] == attr3])#Subset the dataframe based on the attributes

bbq_attribute_imp = bbq_attribute.groupby(['BBQ']).sum()#Summing up the importance for the BBQs

bbq_attribute_imp['imp'].max()#Getting the most relevant BBQ

x = bbq_attribute_imp[bbq_attribute_imp['imp'] == bbq_attribute_imp['imp'].max()]
x = x.reset_index()
x['BBQ'][0]

print('The best BBQ restaurant for the given attributes based on word frequencies is: ' + x['BBQ'][0])
```

### Cosine Similarity

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
def cosine_sim(dataframe,attribute1,attribute2,attribute3):
    attributes = [attribute1,attribute2,attribute3]
    # Create the Document Term Matrix
    count_vectorizer = CountVectorizer(stop_words='english')
    sparse_matrix = count_vectorizer.fit_transform(dataframe['ReviewDescription'])

    # OPTIONAL: Convert Sparse Matrix to Pandas Dataframe if you want to see the word frequencies.
    doc_term_matrix = sparse_matrix.todense()
    df = pd.DataFrame(doc_term_matrix, 
                  columns=count_vectorizer.get_feature_names(),
                     index=dataframe['BBQ'])
    da = pd.DataFrame(columns = attributes)
    da.at[0,attribute1]=1
    da.at[0,attribute2]=1
    da.at[0,attribute3]=1
    df = pd.concat([df,da])
    df = df.fillna(0)
    dj = cosine_similarity(df[:-1].values,np.reshape(df.iloc[-1].values, (-1, len(df.columns))))
    return dj
```

```python
result = cosine_sim(data,'rib','sausage','wait')
```

```python
cos_df = data[['BBQ','ReviewDescription']]
cos_df['similarity_score'] = ''
for lst in range(len(result)):
    for val in result[lst]:
        cos_df["similarity_score"].iloc[lst] = val
```

## Method 2: Sentiment Analysis

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
lemmatizer = WordNetLemmatizer()
vader = SentimentIntensityAnalyzer()
stop_words = set(stopwords.words('english'))
```

```python
def filtered(List):
    list1=[]
    for word in List:
        if word not in stop_words:
            list1.append(word)
    return list1
```

```python
def lemma(List):
    list2 = []
    for word in List:
        list2.append(lemmatizer.lemmatize(word))
    return list2
```

```python
def sentiment(List):
    return vader.polarity_scores(" ".join(List))['compound']
```

```python
data['review_words'] = data['ReviewDescription'].map(lambda x: x.split())
data['filtered_review'] = data['review_words'].map(filtered).map(lemma)
data['sentiment_score'] = data['filtered_review'].map(sentiment)
```

```python
merged_df = data.merge(cos_df,
                   left_on = "ReviewDescription",
                   right_on = "ReviewDescription",
                   suffixes = ("_sentiment",'_similarity'))
```

```python
sentiment = merged_df.groupby(['BBQ_sentiment'])[['sentiment_score']].mean()
sentiment.sort_values(by='sentiment_score',ascending=False)
```

## Method 3: Evaluation Score (similarity+sentiment)

```python
merged_df['similarity_score'] = merged_df['similarity_score'].astype('float')
similarity = merged_df.groupby(['BBQ_sentiment'])[['similarity_score']].mean()
sentiment['similarity_score'] = similarity['similarity_score']
sentiment['evaluation_score'] = sentiment['sentiment_score']+sentiment['similarity_score']
sentiment['evaluation_score'].sort_values(ascending=False)
```

```python
import spacy
import en_core_web_md
from statistics import mean
nlp = en_core_web_md.load()
nlp = spacy.load('en_core_web_md')
```

```python
# Users input their desired attributes
attr1 = nlp(input())
attr2 = nlp(input())
attr3 = nlp(input())
```

```python
df_groupby = data.groupby('BBQ')
bbq_attr1 = []
bbq_attr2 = []
bbq_attr3 = []
wordvector_df = pd.DataFrame(columns = ['BBQ', 'similarity score_attr1', 'similarity score_attr2', 'similarity score_attr3'])

for bbq in bbq_list:
    # create a df with all reviews from each product
    bbq_df = df_groupby.get_group(bbq)
    for row in range(len(bbq_df)):
        doc = nlp(bbq_df.iloc[row]['ReviewDescription'])
        # Calculate similarity with each attribute
        bbq_attr1.append(doc.similarity(attr1))
        bbq_attr2.append(doc.similarity(attr2))
        bbq_attr3.append(doc.similarity(attr3))
    # Calculate average similarity score for each product
    similarity1 = mean(bbq_attr1)
    similarity2 = mean(bbq_attr2)
    similarity3 = mean(bbq_attr3)
    wordvector_df.loc[len(wordvector_df)] = [bbq, similarity1, similarity2, similarity3]
```

```python
# Our Recommendation based on the three attributes:

wordvector_df['mean similarty score'] = wordvector_df[['similarity score_attr1', 'similarity score_attr2', 'similarity score_attr3']].apply(lambda x: x.mean(), axis = 1)
wordvector_df.sort_values(by='mean similarty score', ascending = False)[:3]
```
